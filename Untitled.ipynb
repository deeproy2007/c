```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('netflix_titles.csv')
df.head(21)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['director'],inplace=True)
df.dropna(subset=['cast'],inplace=True)
df.dropna(subset=['country'],inplace=True)
df.dropna(subset=['date_added'],inplace=True)
df.dropna(subset=['rating'],inplace=True)
df.dropna(subset=['duration'],inplace=True)
print(df.isnull().sum())
# df['duration']=df[df['duration'] > 1 ]
df.head(21)
```

```python
import numpy as np
m_v=df['type'].value_counts()
sns.barplot(x=m_v.values,y=m_v.index,palette='viridis')
font1={'color':'green','size':15}
font2={'color':'red','size':20}
plt.title("Movie vs TVshow",fontdict=font2)
plt.xlabel("Count",fontdict=font1)
plt.ylabel("Types",fontdict=font1)
plt.show()
```

```python
lst_a=df['listed_in'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=lst_a.values,y=lst_a.index,palette='cividis')
font1={'color':'red','size':20}
font2={'color':'green','size':15}
plt.title("Top 10 listed Movies",fontdict=font1,loc='left')
plt.xlabel("Count",fontdict=font2)
plt.ylabel("List",fontdict=font2)
plt.show()
```

```python
c_c=df['country'].value_counts().sort_values(ascending=False).head(10)
sns.barplot(x=c_c.values,y=c_c.index,palette='cividis')
font1={'color':'red','size':20}
font2={'color':'green','size':15}
plt.title("Top 10 Country",fontdict=font1)
plt.xlabel("Count",fontdict=font2)
plt.ylabel("Country",fontdict=font2)
plt.show()
```

```python
top_di=df['director'].value_counts().head(10)
sns.barplot(x=top_di.values,y=top_di.index,palette='viridis')
font1={'color':'red','size':20}
font2={'color':'green','size':15}
plt.title("Top 10 Director",fontdict=font1)
plt.xlabel("Count",fontdict=font2)
plt.ylabel("Director",fontdict=font2)
plt.show()
```

```python
movies = df[df['type'] == 'Movie'].copy()

movies['duration_mins'] = movies['duration'].str.replace(' min','').astype(int)

plt.figure(figsize=(10,5))
sns.histplot(movies['duration_mins'], bins=40)
font1={'color':'red','size':20}
font2={'color':'green','size':15}
plt.title("Distribution of Movie Duration (Minutes)",fontdict=font1)
plt.xlabel("Duration (min)",fontdict=font2)
plt.ylabel("Count",fontdict=font2)
plt.show()
```

```python
plt.figure(figsize=(15,10))
tit=df['title'].value_counts().sort_values(ascending=False).head(10)
duration=df['duration'].value_counts().head(10)
sns.barplot(y=tit.index,x=duration.values)
font1={'color':'red','size':25}
font2={'color':'green','size':20}
plt.title("Top 10 Title by Duration",fontdict=font1)
plt.ylabel("Title",fontdict=font2)
plt.xlabel("Duration",fontdict=font2)
plt.show()
```
